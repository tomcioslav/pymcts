```python
from pymcts.games.bridgit.game import BridgitGame
from pymcts.games.bridgit.config import BoardConfig, NeuralNetConfig
from pymcts.games.bridgit.neural_net import BridgitNet
from pymcts.core.config import MCTSConfig, TrainingConfig, ArenaConfig
from pymcts.core.trainer import train
```

## Configure training

```python
board_config = BoardConfig(size=5)
net_config = NeuralNetConfig(num_channels=16, num_res_blocks=2)
mcts_config = MCTSConfig(num_simulations=50)

net = BridgitNet(board_config=board_config, net_config=net_config)

# Print configs for reference
print("Board config:", board_config.model_dump_json(indent=2))
print("Net config:", net_config.model_dump_json(indent=2))
print("MCTS config:", mcts_config.model_dump_json(indent=2))
```

## Run training

```python
train(
    game_factory=lambda: BridgitGame(board_config),
    net=net,
    mcts_config=mcts_config,
    training_config=TrainingConfig(
        num_iterations=3,
        num_self_play_games=10,
    ),
    arena_config=ArenaConfig(
        num_games=10,
        swap_players=True,
    ),
    game_type="bridgit",
    game_config=board_config.model_dump(),
)
```
