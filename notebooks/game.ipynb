```python
from pymcts.games.bridgit.game import BridgitGame, BridgitGameState
from pymcts.games.bridgit.player import Player
from pymcts.games.bridgit.config import BoardConfig
from pymcts.games.bridgit.visualizer import Visualizer
```

### Game

```python
game = BridgitGame()
state = game.get_state()
print(f"Board shape: {state.board.shape}, n={state.n}")
Visualizer.visualize_game_state(state)
```

```python
# Make a move and visualize
game.make_action(game.row_col_to_action(1, 1))
Visualizer.visualize_game_state(game.get_state())
```

```python
# Check the mask (1D boolean: True = legal action)
mask = game.to_mask()
print(f"Mask shape: {mask.shape}, legal moves: {mask.sum().item()}")
print(f"Legal action indices: {game.valid_actions()}")
```

```python
# Visualize the mask as a 2D grid
g = game.board_rows
Visualizer.visualize_array(mask.reshape(g, g).float(), title="Legal moves mask")
```

```python
# Canonicalization demo: state is always from current player's perspective
print(f"Current player: {game.current_player}")
print(f"State is canonical — board transposed for VERTICAL player automatically")
```

```python
# Make more moves
game.make_action(game.row_col_to_action(2, 2))
game.make_action(game.row_col_to_action(3, 3))
print(f"Current player: {game.current_player}, is_over: {game.is_over}")
Visualizer.visualize_game_state(game.get_state())
```
