# Bridgit Training

Self-play → Train → Evaluate loop (AlphaZero-style).

## 1. Configuration

```python
from bridgit.config import Config

config = Config()

# Board
config.board.size = 5

# Neural net
config.neural_net.num_channels = 64
config.neural_net.num_res_blocks = 4

# MCTS
config.mcts.num_simulations = 100
config.mcts.c_puct = 1.5
config.mcts.dirichlet_alpha = 1.0
config.mcts.dirichlet_epsilon = 0.25

# Training
config.training.num_iterations = 50
config.training.num_self_play_games = 50
config.training.num_epochs = 20
config.training.batch_size = 64
config.training.learning_rate = 0.001
config.training.replay_buffer_size = 5

# Arena evaluation
config.arena.num_games = 40
config.arena.threshold = 0.55
config.arena.swap_players = True

print(config.model_dump_json(indent=2))
```

## 2. Train

Set `checkpoint_path` to resume from a previous run, or leave as `None` to start fresh.

```python
from bridgit.ai.trainer import train

checkpoint_path = None  # e.g. "trainings/run_2025-.../iteration_010/post_training.pt"

train(config, checkpoint_path=checkpoint_path)
```
